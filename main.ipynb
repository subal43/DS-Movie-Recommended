```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
imp_fea_data = imp_fea_data.copy()
imp_fea_data['genres'] = imp_fea_data['genres'].apply(convert)
```

```python
imp_fea_data['keywords'] = imp_fea_data['keywords'].apply(convert)
```

```python
imp_fea_data.iloc[0,5]
```

```python
imp_fea_data['production_companies'] = imp_fea_data['production_companies'].apply(convert)
```

```python
def convert2(obj):
    L = []
    c = 0
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        c += 1
        if c>2:
            break
    return L
```

```python
imp_fea_data['cast'].apply(convert2).iloc[0]
```

```python
imp_fea_data['cast'] = imp_fea_data['cast'].apply(convert2)
```

```python
imp_fea_data.head(1)
```

```python
def convert3(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
imp_fea_data['crew'] = imp_fea_data['crew'].apply(convert3)
```
