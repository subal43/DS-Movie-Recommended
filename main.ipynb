```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert2(obj):
    L = []
    c = 0
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        c += 1
        if c>2:
            break
    return L
```

```python
imp_fea_data['cast'].apply(convert2).iloc[0]
```

```python
def convert3(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
imp_fea_data['tag'] = imp_fea_data['genres']+imp_fea_data['keywords']+imp_fea_data['overview']+imp_fea_data['production_companies']+imp_fea_data['cast']+imp_fea_data['crew']
```

```python
new_df = imp_fea_data[['id','title','tag']].copy()
```

```python
new_df['tag']
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x : " ".join(x))
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x : x.lower())
```

```python
new_df.head(1)
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stemming(text):
    s = []
    for i in text.split():
        s.append(ps.stem(i))
    return ' '.join(s)
```

```python
new_df['tag'] = new_df['tag'].apply(stemming)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000 , stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
cv.get_feature_names_out()[0:20]
```

```python
vectors.shape
```

```python
new_df.head()
```

```python
new_df[new_df['title'] == 'The Dark Knight Rises'].index[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    movie_list = sorted(list(enumerate(similarity[movie_index])),reverse=True , key = lambda x : x[1])[1:6]
    for i in movie_list :
        print(fetch_poster(new_df.iloc[i[0]].id))
        
        
        
        
    
```

```python
import requests
def fetch_poster(movie_id):
    url = f"https://api.themoviedb.org/3/movie/{movie_id}?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US"
    data = requests.get(url).json()
    # print(data)
    if data.get('poster_path'):
        return "https://image.tmdb.org/t/p/w500/" + data['poster_path']
    else:
        return "https://via.placeholder.com/300x450?text=No+Image"
```
